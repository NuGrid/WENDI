## 1.) First step : Running the simulation

Start the simulation by executing the following line. You create a single stellar population out of a gas cloud of $10^{11}M_{\odot}$ and a metallicity of $Z=10^{-4}$ ([Fe/H]=-2.3) and
let it evolve over $10^{9}$ years. AGB, massive stars with SN2 as well as SN1a eject their material into the ISM.

```python
s1=s.sygma()
```

Thats it! Now you can follow the evolution of all elements and many isotopes by using SYGMA's functions.

## 2) Visualization

Now lets plot the evolution of the total gas mass in the interstellar medium (ISM) in $M_{\odot}$ over time with the following command.

```python
s1.plot_totmasses(source='all')
```

'All' indicates that you sum all sources which eject into the ISM: AGB, massive and SNIa. Keep in mind that much gas is locked away in the remnants.
Now lets see what amount only AGB eject into the ISM (set by using the source variable):

```python
s1.plot_totmasses(source='agb')
```

AGB stars conribute not from the beginning but later during the chemical evolution as you can see from the diagram.

In order to compare also massive and SN1a contribution you can do many plotting commands in one line.
Each graph has its own marker, color and label as an identifier.

```python
s1.plot_totmasses(source='agb')
s1.plot_totmasses(source='massive')
s1.plot_totmasses(source='sn1a')
```

```python
s1.plot_mass(specie='Fe',source='massive')
s1.plot_mass(specie='Fe',source='all')
```

## 3) Data extraction

To exract their formation about chemical evolution, yield tables in solar masses of elements and isotopes can be downloaded. Here the elements H, He and C are extracted and can be downloaded with the link.

```python
s1.write_evol_table(elements=['H','He','C'])
```

Each lines represents a timestep with the ISM composition of the species.
